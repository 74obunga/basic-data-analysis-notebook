```python
#A z-score is a measurement that describes the distance a value is from the mean of a normal data distribution. Z-score is measured in terms of standard deviations. A z-score typically tells us how far a value is in standard deviations, so that we can understand it in a scaled value.

#A z-score can be calculated using the following formula: z = (X – μ) / σ

#X is the single input value

#μ is the population mean

#σ is the population standard deviation
```

```python
# Lets find for instance our student dataset, the z-scores for OfficeHoursParticipated values.

mean = grades_1['OfficeHoursParticipated'].mean()
std = grades_1['OfficeHoursParticipated'].std()
```

```python
grades_1['z-sscore'] = (grades_1['OfficeHoursParticipated'] - mean)/std
```

# Identifying outliers


```python
#Outliers are defined as data points that are abnormally far from what is considered "normal".
#Sometimes, outliers causes hinderance to our data, as they might be there due to errors and it is up to us to decide if these abnormal data are to be included. 
#Before we start identifying outliers, we are going to add a new column to our DataFrame GPA, based on the GradeAverage of the students. We will create this using the np.select function.

```

```python

grades_1['GPA'] = np.select([grades_1['GradeAverage'] == 'A', 
                             grades_1['GradeAverage'] == 'B',
                             grades_1['GradeAverage'] == 'C',
                             grades_1['GradeAverage'] == 'D',
                             grades_1['GradeAverage'] == 'F'],
                            [4,3,2,1,0])
grades_1                       
```

```python
#Then e calculate the interqutile range for GPA scores.
q3 = np.percentile(grades_1['GPA'], 75)
q1 = np.percentile(grades_1['GPA'], 25)
IQR = q3-q1
IQR
```

```python
# Define IQR rule for the GPA column
IQR_rule = (1.5*IQR + q3 < grades_1['GPA']) | (q1 - 1.5*IQR > grades_1['GPA'])

# use np.where to create 'Outlier' column, is a student's GPA an outlier or not
grades_1['Outlier'] = np.where(IQR_rule, 'yes', 'no')
grades_1
```
